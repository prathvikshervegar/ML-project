```python
# importing libraries    
import sys
import numpy as nm    
import matplotlib.pyplot as mtp    
import pandas as pd   
```

```python
# Importing the dataset  
dataset = pd.read_csv('Mall_Customers.csv')
dataset.head(10)
```

```python
#total rows and columns in the dataset
dataset.shape
```

```python
# 1) Considering Age and Spending score
x1= dataset[['Age' , 'Spending Score (1-100)']].iloc[:, :].values
```

```python
#Finding optimal number of clusters using the elbow method for the above
from sklearn.cluster import KMeans  
wcss_list= []  #Initializing the list for the values of WCSS  
  
#We assume max number of clusters would be 10
#by doing average we can judge the number of clusters
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
    kmeans.fit(x1)  
    wcss_list.append(kmeans.inertia_)  
```

```python
#Visualizing the ELBOW method to get the optimal value of k
mtp.plot(range(1, 11), wcss_list)  
mtp.title('The Elbow Method Graph')  
mtp.xlabel('Number of clusters(k)')  
mtp.ylabel('WCSS')  
mtp.show() 
```

```python
#From the curve we will see that last elbow comes at k=4
#Therefore, k=4

#training the K-means model on a dataset  
kmeans = KMeans(n_clusters=4, init='k-means++', random_state= 42)  
y_predict= kmeans.fit_predict(x1)

#y_predict is the final model. Now how we deploy this model will depend on tool we use
```

```python
#visulaizing the clusters  
mtp.scatter(x1[y_predict == 0, 0], x1[y_predict == 0, 1], s = 100, c = 'blue', label = 'Cluster 1') #for first cluster  
mtp.scatter(x1[y_predict == 1, 0], x1[y_predict == 1, 1], s = 100, c = 'green', label = 'Cluster 2') #for second cluster  
mtp.scatter(x1[y_predict== 2, 0], x1[y_predict == 2, 1], s = 100, c = 'red', label = 'Cluster 3') #for third cluster  
mtp.scatter(x1[y_predict == 3, 0], x1[y_predict == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4') #for fourth cluster   
mtp.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')   
mtp.title('Clusters of customers')  
mtp.xlabel('Age')  
mtp.ylabel('Spending Score (1-100)')  
mtp.legend()  
mtp.show()  
```

```python
# 2) Considering Annual income and Spending score
x2= dataset[['Annual Income (k$)' , 'Spending Score (1-100)']].iloc[:, :].values
```

```python
#Finding optimal number of clusters using the elbow method
from sklearn.cluster import KMeans  
wcss_list= []  #Initializing the list for the values of WCSS  
  
#We assume max number of clusters would be 10
#by doing average we can judge the number of clusters
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
    kmeans.fit(x2)  
    wcss_list.append(kmeans.inertia_)  
```

```python
#Visualizing the ELBOW method to get the optimal value of k
mtp.plot(range(1, 11), wcss_list)  
mtp.title('The Elbow Method Graph')  
mtp.xlabel('Number of clusters(k)')  
mtp.ylabel('WCSS')  
mtp.show()
```

```python
#From the curve we will see that last elbow comes at k=5
#Therefore, k=5

#training the K-means model on a dataset  
kmeans = KMeans(n_clusters=5, init='k-means++', random_state= 42)  
y_predict= kmeans.fit_predict(x2)

#y_predict is the final model. Now how we deploy this model will depend on tool we use
```

```python
#visulaizing the clusters  
mtp.scatter(x2[y_predict == 0, 0], x2[y_predict == 0, 1], s = 100, c = 'blue', label = 'Cluster 1') #for first cluster  
mtp.scatter(x2[y_predict == 1, 0], x2[y_predict == 1, 1], s = 100, c = 'green', label = 'Cluster 2') #for second cluster  
mtp.scatter(x2[y_predict== 2, 0], x2[y_predict == 2, 1], s = 100, c = 'red', label = 'Cluster 3') #for third cluster  
mtp.scatter(x2[y_predict == 3, 0], x2[y_predict == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4') #for fourth cluster  
mtp.scatter(x2[y_predict == 4, 0], x2[y_predict == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5') #for fifth cluster  
mtp.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')   
mtp.title('Clusters of customers')  
mtp.xlabel('Annual Income (k$)')  
mtp.ylabel('Spending Score (1-100)')  
mtp.legend()  
mtp.show()  
```
